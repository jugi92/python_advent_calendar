```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
```

The cell below defines two functions x and y that are sine and cosine curves with shrinking amplitudes.

```python
theta = np.linspace(-8 * np.pi, 8 * np.pi, 200)
z = np.linspace(-3, 0, 200)
r = 5
x = r * np.sin(theta) * z
y = r * np.cos(theta) * z
```

```python
plt.plot(x)
```

```python
plt.plot(y)
```

```python
plt.plot(z)
```

```python
# Combined they create a 3D spiral starting from point 0, 0, 0.
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.view_init(15, 0)  # Set a nice view angle
ax.plot(x, y, z)
```

```python
ax.plot(x, y, z, c='green', linewidth=2.5)
fig
```
